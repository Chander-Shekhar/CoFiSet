```python
import numpy as np 
import pandas as pd

data = pd.read_table('../dataset/udata.txt', delim_whitespace=True, header = None, parse_dates=False)
dataset = data.values[:, :2]
feedback = data.values[:, 2]
observed_data = dataset[feedback > 3]
users = np.unique(observed_data[:, 0])
maxi = max(observed_data[:, 1])
train = []
test = []
tmp = []

for user in range(942):
    indices = (np.where(observed_data[:, 0] == users[user]))[0]
    tmp = observed_data[indices][:, 1]
    train_data = np.random.choice(tmp, size=tmp.shape[0]//2, replace=False)
    train.append(train_data - np.ones(train_data.shape[0], dtype=int))
    test.append(np.array([item - 1 for item in tmp if item not in list(train_data)]))
```

```python
# Data Preprocessing
#data = pd.read_table('../dataset/u1base.txt', delim_whitespace=True, header = None, parse_dates=False)
#dataset=data.values[:,:2]
#feedback=data.values[:,2]
#users=data.values[:,0]
d=10

# maxi=max(train[0])
# for i in range(1,len(train)):
#     if(maxi<max(train[i])):
#         maxi=max(train[i])

#Data Defining
Users = np.zeros((len(train), d), dtype=np.float32)
Items= np.zeros((maxi, d), dtype=np.float32)
Bias= np.zeros(maxi, dtype=np.float32)

P_users=[]
A_users=[]
total_items=np.array(range(maxi))
for u in range(Users.shape[0]):
    P_users.append(np.array(train[u]))
    A_users.append(np.array([item for item in list(total_items) if item not in list(P_users[u])]))
    
#Initialization
for u in range(Users.shape[0]):
    Users[u] = (np.random.uniform(0, 1, d) - 0.5) * 0.1
    
for i in range(Items.shape[0]):
    Items[i] = (np.random.uniform(0, 1, d) - 0.5) * 0.1

for i in range(Items.shape[0]):
    b = 0
    for u in range(Users.shape[0]):
        if(i in P_users[u]):
            b += A_users[u].shape[0]
    Bias[i] = b/(Items.shape[0] * Users.shape[0])

T=10
learning_rate=0.01
alpha_u=alpha_v=beta_v=0.01

print(type(Users[0][0]), type(Items[0][0]), type(Bias[0]))
```

```python
def sigmoid_deriv(x):
    sigm = 1. / (1. + np.exp(-x))
    return sigm * (1. - sigm)


def sigmoid(x):
    sigm = 1. / (1. + np.exp(-x))
    return sigm

def preference(user,P,A):
    rp=0
    for i in range(P.shape[0]):
        rp+=np.dot(Users[user],Items[P[i]])+Bias[P[i]]
        
    ra=0
    for i in range(A.shape[0]):
        ra+=np.dot(Users[user],Items[A[i]])+Bias[A[i]]
    
    return rp/P.shape[0]-ra/A.shape[0]

def calc_mean(d,P,A):
    P_mean=np.zeros((d))
    for i in P:
        P_mean+=Items[i]
    A_mean=np.zeros((d))
    for i in A:
        A_mean+=Items[i]
    P_mean=P_mean/P.shape[0]
    A_mean=A_mean/A.shape[0]
    return P_mean,A_mean


    
```

```python
for t1 in range(2000):
    loss=0
    for user in range(Users.shape[0]):
        
        
        # Experiment with the size of P and A.
        
        
        p_size=np.random.randint(d)+1
        a_size=np.random.randint(p_size)+1
#         P_indices=np.random.randint(P_users[user].shape[0],size=p_size)
#         P=P_users[user][P_indices]
#         A_indices=np.random.randint(int(len(A_users[user])),size=a_size)
#         A=A_users[user][A_indices]
        sizeP = min(P_users[user].shape[0], p_size)
        sizeA = min(A_users[user].shape[0], a_size)
        P = np.random.choice(P_users[user], size=sizeP, replace=False)
        A = np.random.choice(A_users[user], size=sizeA, replace=False)
        rpa=preference(user,P,A)
        #loss
        loss+=-np.log(sigmoid(rpa))
        #derivate of loss fn with respect to R_upa
        
        del_rpa=-sigmoid(-rpa)
        P_mean,A_mean=calc_mean(d,P,A)
        
        
        
        #update equation
        Users[user]-= learning_rate*((P_mean-A_mean)*del_rpa + alpha_u*Users[user])
        for i in P:
            Items[i]-= learning_rate*(del_rpa*Users[user]/P.shape[0] + alpha_v*Items[i])
            
#         P_mean,A_mean=calc_mean(d,P,A)
#         Users[user]-= learning_rate*((P_mean-A_mean)*del_rpa + alpha_u*Users[user])
        
        for i in A:
            Items[i]-= learning_rate*(-del_rpa*Users[user]/A.shape[0] + alpha_v*Items[i])
        
#         P_mean,A_mean=calc_mean(d,P,A)
#         Users[user]-= learning_rate*((P_mean-A_mean)*del_rpa + alpha_u*Users[user])
        
        for i in P:
            Bias[i]-= learning_rate*(del_rpa/P.shape[0] + beta_v*Bias[i] )
        for i in A:
            Bias[i]-= learning_rate*(del_rpa/A.shape[0] + beta_v*Bias[i] )
    loss_epoch=loss/Users.shape[0]
    print(t1,loss_epoch)
```
